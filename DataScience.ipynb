```python
def treat_missing_values(dataframe, main_name, column_to_fill, multiplier = 1000):
  main_column =  dataframe.loc[:,[main_name, column_to_fill]]
  treatet_row = main_column.groupby(main_name).cumcount() + 1
  main_column[column_to_fill] = treatet_row
  main_column[column_to_fill] = main_column[main_name] * multiplier + main_column[column_to_fill]
  
  dataframe[column_to_fill] = main_column[column_to_fill]
  return dataframe

```
